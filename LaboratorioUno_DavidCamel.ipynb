### Enfoque de escala de grises ponderado
##### Método promedio
El método promedio, utiliza el promeido de los tres colores. Dado que es una imagen RGB, significa que debe agregar r con g con b y luego dividirlo por 3 para obtener la imagen en escala de grises deseada.
##### Método ponderado o método de luminosidad (BT.601)
Dado que el color rojo tiene más longitud de onda de los tres colores, y el verde es el color que no solo tiene menos longitud de onda que el color rojo, sino que también el verde es el color que le da un efecto más relajante a los ojos.
Significa que tenemos que disminuir la contribución del color rojo y aumentar la contribución del color verde, y poner la contribución del color azul entre estos dos.
Nueva imagen en escala de grises = ((0.3 * R) + (0.59 * G) + (0.11 * B)).
##### Estandar BT 709
El estándar BT.709 tiene la siguiente fórmula Y= 0.212R + 0.715G + 0.0722B. Cómo se puede observar, esté método le da mayor ponderación al canal verde. Esto se debe a que el ojo humano le da más importancia al color verde


```python
def ConverGris(ImG_Gris, MetodoGris):
    Xn = int(MetodoGris)               
    img_G = cv2.imread(ImG_Gris)
    img_G1 = cv2.cvtColor(img_G, cv2.COLOR_BGR2RGB) #CONVERT COLOR
   # plt.imshow(img_G1) # lee en BGR
   # plt.show()
    
    alto  = img_G1.shape[0]
    ancho = img_G1.shape[1]
    NuevaImg_G1 = np.zeros((alto, ancho,3))
    #1: Promedio # 2: BT.601 # 3: BT.709
    if Xn == 1:
            for i in range (0,alto):#alto
                for j in range (0,ancho):#ancho
                    NuevaImg_G1[i,j] =  int(img_G1[i,j,0]/3 + img_G1[i,j,1]/3 + img_G1[i,j,2]/3) 
    if Xn == 2:
            for i in range (0,alto):#alto
                for j in range (0,ancho):#ancho
                    NuevaImg_G1[i,j] =  int(img_G1[i,j,0]*0.114 + img_G1[i,j,1]*0.587 + img_G1[i,j,2]*0.299) 
    if Xn == 3:
            for i in range (0,alto):#alto
                for j in range (0,ancho):#ancho
                    NuevaImg_G1[i,j] =  int(img_G1[i,j,0]*0.0722 + img_G1[i,j,1]*0.715 + img_G1[i,j,2]*0.212) 
   

        
    cv2.imwrite("NuevoGris.jpg", NuevaImg_G1)
    NEW_G = cv2.imread("NuevoGris.jpg")
    monocromo = cv2.cvtColor(NEW_G, cv2.COLOR_BGR2RGB)
    plt.imshow(monocromo)
    plt.show()

     
```

### En que consiste el espacio de color HSV

En el modelo de color HSV, un color se define por su tono (H), su saturación (S) y su luminosidad (V). A diferencia del modelo RGB ampliamente usado en los monitores, televisores. El color HSV sigue una representación más parecida a las coordenadas cilíndricas. Además es una representación más cercana a la forma en que los humanos perciben los colores y sus propiedades, pues se agrupan las tonalidades de color, lo cual es distinto al caso RGB donde los colores no están necesariamente tan agrupados.

##### El tono (H) 
se da como un ángulo en el círculo cromático, por lo tanto, puede alcanzar valores entre 0° y 360°. 
        * 0° corresponde al color rojo, 
        * 120° corresponde al color verde y 
        * 240° corresponde al color azul. 
##### La saturación (S) 
se declara como porcentajes y, por lo tanto, puede alcanzar valores entre 0% y 100% (o 0 a 1). Una saturación del 100% significa un color completamente saturado y puro, cuanto menor es la saturación, más se vuelve el color a un gris neutro. 
##### luminosidad o negrura (V)
también se da como un porcentaje, donde 0% significa que no hay brillo (por lo tanto, negro) y 100% de brillo total, por lo tanto, un espectro entre el color puro (saturación del 100%) y el blanco (saturación de 0 %)

Si ambos, la saturación y la luminosidad son 100%, se obtiene un color puro. Si la saturación es 0% y la luminosidad es 100%, es blanca y para todos los casos en que la luminosidad es 0%, es negra.
